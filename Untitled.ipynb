```python
import pandas as pd

df = pd.read_excel(r"C:\Users\LENOVO\OneDrive\Documents\pemodelan dan simulasi\modsim-2026-p2-ifs25026-latihan\data_kuesioner.xlsx")

# ambil kolom Q1 sampai Q17
kolom_q = [f"Q{i}" for i in range(1, 18)]

# hitung jumlah TS di tiap pertanyaan
jumlah_ts = (df[kolom_q] == "TS").sum()

# hitung persen TS
persen_ts = (df[kolom_q] == "TS").mean() * 100

# gabungkan hasil
hasil = pd.DataFrame({
    "Jumlah TS": jumlah_ts,
    "Persen TS (%)": persen_ts
})

# cari pertanyaan dengan TS terbanyak
terbanyak = hasil.loc[hasil["Jumlah TS"].idxmax()]

print("Pertanyaan dengan TS terbanyak:")
print(terbanyak)

print("\nDetail semua pertanyaan:")
print(hasil.sort_values(by="Jumlah TS", ascending=False))
```
