# Ploting dengan Matplotlib

## 0. Import Library

```python
import matplotlib.pyplot as plt
import numpy as np
```


## 1. Line Plot (Grafik Garis)

Digunakan untuk melihat tren atau perubahan data terhadap waktu atau urutan tertentu.

```python
# Data
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun']
penjualan = [20, 25, 30, 28, 35, 40]

# Membuat line plot
plt.plot(bulan, penjualan)

# Memberi judul dan label
plt.title('Pertumbuhan Penjualan Per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan (Juta Rupiah)')

# Menampilkan grafik
plt.show()
```

## 2. Scatter Plot (Grafik Sebar)

Digunakan untuk menampilkan hubungan atau korelasi antara dua variabel.

```python

# Data
jam_belajar = [1, 2, 3, 4, 5, 6, 7, 8]
nilai = [60, 65, 70, 75, 80, 85, 90, 95]

# Membuat scatter plot
plt.scatter(jam_belajar, nilai)

# Memberi judul dan label
plt.title('Hubungan Jam Belajar dan Nilai Ujian')
plt.xlabel('Jam Belajar (jam)')
plt.ylabel('Nilai Ujian')

# Menampilkan grafik
plt.show()
```


## 3. Bar Plot (Grafik Batang)

Digunakan untuk membandingkan nilai antar kategori.

```python

# Data
jurusan = ['S1IF', 'S1SI', 'S1TE', 'S1TM', 'S1MR']
jumlah_mahasiswa = [120, 95, 80, 60, 70]

# Membuat bar plot
plt.bar(jurusan, jumlah_mahasiswa)

# Memberi judul dan label
plt.title('Jumlah Mahasiswa per Jurusan')
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')

# Menampilkan grafik
plt.show()
```

## 4. Histogram

Digunakan untuk menunjukkan distribusi frekuensi data numerik.

```python

# Data nilai ujian
nilai_ujian = [
    60, 65, 70, 72, 75, 78, 80, 82, 85, 88,
    90, 92, 95, 70, 68, 74, 76, 84, 86, 89
]

# Membuat histogram
plt.hist(nilai_ujian, bins=5)

# Memberi judul dan label
plt.title('Distribusi Nilai Ujian Mahasiswa')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')

# Menampilkan grafik
plt.show()
```

## 5. Pie Chart (Diagram Lingkaran)

Digunakan untuk menampilkan proporsi atau persentase dari keseluruhan.

```python
# Data
platform = ['Android', 'iOS', 'Web']
jumlah_pengguna = [500, 300, 200]

# Membuat pie chart
plt.pie(jumlah_pengguna, labels=platform, autopct='%1.1f%%')

# Judul
plt.title('Persentase Pengguna Aplikasi')

# Menampilkan grafik
plt.show()
```

## 6. Box Plot

Digunakan untuk melihat sebaran data, median, dan outlier.
Contoh: analisis gaji karyawan.

```python
# Data gaji karyawan (juta rupiah)
gaji = [5, 6, 6, 7, 7, 8, 8, 9, 10, 12, 15, 30]

# Membuat box plot
plt.boxplot(gaji)

# Judul dan label
plt.title('Box Plot Gaji Karyawan')
plt.ylabel('Gaji (Juta Rupiah)')

# Menampilkan grafik
plt.show()
```


## 7. Area Plot

Area Plot digunakan untuk menampilkan perubahan nilai data dari waktu ke waktu atau antar kategori, dengan menekankan besarnya kontribusi (volume) suatu nilai melalui luas area di bawah grafik.

```python
# Data
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun']
pengguna_kumulatif = [100, 180, 260, 350, 450, 600]

# Membuat area plot
plt.fill_between(bulan, pengguna_kumulatif)

# Judul dan label
plt.title('Pertumbuhan Kumulatif Pengguna Aplikasi')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Pengguna')

# Menampilkan grafik
plt.show()
```

## 8. Stack Plot

Digunakan untuk menampilkan kontribusi beberapa data terhadap total.

```python
# Data
tahun = [2019, 2020, 2021, 2022, 2023]
energi_surya = [20, 25, 30, 40, 50]
energi_angin = [30, 35, 40, 45, 55]
energi_air = [50, 55, 60, 65, 70]

# Membuat stack plot
plt.stackplot(
    tahun,
    energi_surya,
    energi_angin,
    energi_air,
    labels=['Surya', 'Angin', 'Air']
)

# Judul dan label
plt.title('Kontribusi Sumber Energi dari Waktu ke Waktu')
plt.xlabel('Tahun')
plt.ylabel('Produksi Energi (GWh)')
plt.legend()

# Menampilkan grafik
plt.show()
```


## 9. Step Plot

Digunakan untuk menampilkan data diskrit dengan perubahan nilai yang terjadi pada titik-titik tertentu secara jelas.

```python

# Data
jam = [8, 9, 10, 11, 12, 13]
status_mesin = [0, 0, 1, 1, 0, 1]

# Membuat step plot
plt.step(jam, status_mesin, where='post')

# Judul dan label
plt.title('Perubahan Status Mesin')
plt.xlabel('Jam')
plt.ylabel('Status Mesin (0=Mati, 1=Menyala)')

# Menampilkan grafik
plt.show()
```

## 10. Error Bar Plot

Digunakan untuk menampilkan nilai data beserta tingkat kesalahan atau ketidakpastiannya agar analisis lebih akurat dan dapat dipercaya.

```python
# Data
percobaan = [1, 2, 3, 4, 5]
suhu = [30, 35, 40, 45, 50]
error = [2, 1.5, 2, 1, 2.5]

# Membuat error bar plot
plt.errorbar(percobaan, suhu, yerr=error, fmt='o')

# Judul dan label
plt.title('Hasil Eksperimen dengan Error Bar')
plt.xlabel('Percobaan')
plt.ylabel('Suhu (C)')

# Menampilkan grafik
plt.show()
```

## 11. Stem Plot

Digunakan untuk menampilkan data diskrit secara sederhana, terutama untuk melihat pola, tren, dan perubahan nilai pada data seperti sinyal digital atau data berurutan, dengan menekankan titik data individual.

```python
# Data
waktu = [0, 1, 2, 3, 4, 5, 6]
amplitudo = [0, 2, 1, 3, 2, 4, 3]

# Membuat stem plot
plt.stem(waktu, amplitudo)

# Judul dan label
plt.title('Stem Plot Sinyal Digital')
plt.xlabel('Waktu')
plt.ylabel('Amplitudo')

# Menampilkan grafik
plt.show()
```


## 12. Heatmap

Digunakan untuk menampilkan pola, hubungan, atau tingkat intensitas data secara visual menggunakan warna, sehingga memudahkan analisis korelasi, perbandingan nilai, dan identifikasi pola dalam dataset yang kompleks.

```python
# Data korelasi
korelasi = np.array([
    [1.0, 0.8, 0.6, 0.4],
    [0.8, 1.0, 0.7, 0.5],
    [0.6, 0.7, 1.0, 0.6],
    [0.4, 0.5, 0.6, 1.0]
])

label = ['Matematika', 'Fisika', 'Kimia', 'Biologi']

# Membuat heatmap
plt.imshow(korelasi)
plt.colorbar()

# Label sumbu
plt.xticks(range(len(label)), label)
plt.yticks(range(len(label)), label)

# Judul
plt.title('Heatmap Korelasi Mata Pelajaran')

# Menampilkan grafik
plt.show()
```
